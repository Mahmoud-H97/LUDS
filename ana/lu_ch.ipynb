```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# File paths
pth_2019 = "/tudelft.net/staff-umbrella/EDT Veluwe/testbed/lgntest/LGN2020.tif"
pth_2018 = "/tudelft.net/staff-umbrella/EDT Veluwe/testbed/lgntest/LGN2021.tif"
output_path = "/tudelft.net/staff-umbrella/EDT Veluwe/testbed/lgntest/LGN_change_2020_2021.tif"

# Open the datasets
with rasterio.open(pth_2019) as src_19, rasterio.open(pth_2018) as src_18:

    # Ensure dimensions match
    if src_19.shape != src_18.shape:
        raise ValueError("The rasters do not match in shape!")

    # Copy profile and update for the output
    profile = src_19.profile
    profile.update(dtype=rasterio.uint8, count=1, compress='lzw')

    # Create output raster
    with rasterio.open(output_path, 'w', **profile) as dst:

        # Use tqdm to track progress
        for row in tqdm(range(src_19.height), desc="Comparing rows"):
            # Define the window for the current row
            window = rasterio.windows.Window(col_off=0, row_off=row, width=src_19.width, height=1)

            # Read one row from both rasters
            row_19 = src_19.read(1, window=window)
            row_18 = src_18.read(1, window=window)

            # Compare values for change detection (1 = change, 0 = no change)
            row_change = np.where(row_19 != row_18, 1, 2).astype(np.uint8)

            # Write the result row to the output raster
            dst.write(row_change, 1, window=window)

# Optional: read and plot the result
'''
with rasterio.open(output_path) as src:
    change_map = src.read(1)

plt.figure(figsize=(10, 8))
plt.imshow(change_map, cmap='gray')
plt.title('Land Use Change (2018  2019)')
plt.colorbar(label='1 = Changed, 0 = No Change')
plt.axis('off')
plt.show()
'''
```
